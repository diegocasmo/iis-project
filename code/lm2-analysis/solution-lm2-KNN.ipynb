# KNN Classifier 
KNN classifier on lm2 dataset. Remember to check path of lm3.csv before executing the file

```python
# Load original dataset
import pandas as pd

file_path = r'lm2.csv'
data_df = pd.read_csv(file_path)
print('Num of features: %s' % len(data_df.columns.values))
data_df.head()
```

```python
from preprocess_lm2 import preprocess_data, reduce_features

column_names, principal_labels_df = reduce_features(preprocess_data(data_df))
```

```python
from sklearn import neighbors
from sklearn.metrics import accuracy_score
from constants import get_all_emotions
from splitting import get_train_test_split

runs = 100
knn_scores = []
for index in range(runs):
    # Split train and test labels/data
    #train_df,test_df,train_data,train_labels,test_data,test_labels = get_train_test_split(principal_labels_df,column_names)
    train_data,train_labels,test_data,test_labels = get_train_test_split(principal_labels_df,column_names,index)
    
    # Use KNN to predict
    kNNClassifier = neighbors.KNeighborsClassifier(n_neighbors=10).fit(train_data, train_labels)
    predicted_labels = kNNClassifier.predict(test_data)
    acc_knn = accuracy_score(test_labels, predicted_labels)
    knn_scores.append(acc_knn)
 
print('--------------------Result--------------------')
print("kNN classifier accuracy: %.02f%%" % (sum(knn_scores) / runs)) 
```
