# Random Forest
Random Forest classifier on lm3 dataset. Remember to check path of lm3.csv before executing the file

```python
# Load original dataset
import pandas as pd

file_path = r'../../data/lm3.csv'
data_df = pd.read_csv(file_path)
print('Num of features: %s' % len(data_df.columns.values))
data_df.head()
```

```python
from preprocess_lm3 import preprocess_data, reduce_features

column_names, principal_labels_df = reduce_features(preprocess_data(data_df))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from constants import get_all_emotions
from splitting import get_train_test_split
from numpy import std

runs = 100
random_scores = []
for index in range(runs):
    # Split train and test labels/data
    train_data,train_labels,test_data,test_labels = get_train_test_split(principal_labels_df,column_names,index)
    
    # Use Random Forest to predict
    clf = RandomForestClassifier(n_estimators=200, n_jobs=1,criterion='gini', 
                                 oob_score = True, max_depth=10, 
                                 max_features='sqrt', min_samples_leaf = 1)
    # Train the model on training data
    clf.fit(train_data, train_labels)
    random_predict_labels = clf.predict(test_data)
    random_acc = accuracy_score(test_labels, random_predict_labels)
    random_scores.append(random_acc)

print('--------------------Result--------------------')
print("Random Forest classifier accuracy: %.04f%% SD: %.04f" % (sum(random_scores) / runs, std(random_scores)))
```
