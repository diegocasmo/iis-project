# MLP Classifier
MLP classifier on lm3 dataset. Remember to check path of lm3.csv before executing the file

```python
# Load original dataset
import pandas as pd

file_path = r'lm3.csv'
data_df = pd.read_csv(file_path)
print('Num of features: %s' % len(data_df.columns.values))
data_df.head()
```

```python
from preprocess_lm3 import preprocess_data, reduce_features

column_names, principal_labels_df = reduce_features(preprocess_data(data_df))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from constants import get_all_emotions
from splitting import get_train_test_split

runs = 100
mlp_scores = []
for index in range(runs):
    # Split train and test labels/data
    train_data,train_labels,test_data,test_labels = get_train_test_split(principal_labels_df,column_names,index)
    
    # MLP classifier to predict
    mlp = MLPClassifier(solver='adam', activation='relu', \
                        hidden_layer_sizes=(21), max_iter=200)
    mlp.fit(train_data, train_labels)
    predicted_labels = mlp.predict(test_data)
    acc_mlp = accuracy_score(test_labels, predicted_labels)
    mlp_scores.append(acc_mlp)

print('--------------------Result--------------------')
print('MLP classifier accuracy: %.02f%%' % (sum(mlp_scores) / runs))   
```
